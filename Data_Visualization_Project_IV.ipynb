```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Penguins Data.csv")
df.head(20)
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Culmen Length (mm)"].dropna(), bins=np.arange(30, 60+2, 2), color='skyblue', edgecolor='black', linewidth=1.2)
plt.title("Culmen Length (mm) Histogram", fontsize=14, fontweight='bold')
plt.xlabel("Culmen Length (mm)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Culmen Depth (mm)"].dropna(), bins=np.arange(12, 22+1, 1), color='salmon', edgecolor='black', linewidth=1.2)
plt.title("Culmen Depth (mm) Histogram", fontsize=14, fontweight='bold')
plt.xlabel("Culmen Depth (mm)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Flipper Length (mm)"].dropna(), bins=np.arange(170, 240+5, 5), color='lightgreen', edgecolor='black', linewidth=1.2)
plt.title("Flipper Length (mm) Histogram", fontsize=14, fontweight='bold')
plt.xlabel("Flipper Length (mm)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Body Mass (g)"].dropna(), bins=np.arange(2500, 6500+200, 200), color='violet', edgecolor='black', linewidth=1.2)
plt.title("Body Mass (g) Histogram", fontsize=14, fontweight='bold')
plt.xlabel("Body Mass (g)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
