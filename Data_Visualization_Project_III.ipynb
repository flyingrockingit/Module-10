```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
day_jul = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
new_births_jul = [12, 15, 11, 9, 1, 9, 21]

plt.figure(figsize=(8, 5))
plt.bar(day_jul, new_births_jul, color='orange', edgecolor='black', linewidth=1.5, label='Number of Births')
plt.xlabel("Day of the Week")
plt.ylabel("Number of Births")
plt.title("Birth Rate in July")
plt.grid(axis='y', linestyle=':', linewidth=0.8)
plt.legend()
plt.show()
```

```python
day_aug = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
new_births_aug = [17, 5, 2, 11, 1, 8, 29]

plt.figure(figsize=(8, 5))
plt.bar(day_aug, new_births_aug, color='cyan', edgecolor='black', linewidth=1.5, label='New Births')
plt.xlabel("Day of the Week")
plt.ylabel("Number of Births")
plt.title("Birth Rate in August")
plt.grid(axis='y', linestyle=':', linewidth=0.8)
plt.legend()
plt.show()
```

```python
days = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday", "Sunday"]
births_july = [12, 15, 11, 9, 1, 9, 21]
births_august = [17, 5, 2, 11, 1, 8, 29]
x = np.arange(len(days))
width = 0.35
plt.figure(figsize=(10, 6))
plt.bar(x - width/2, births_july, width,color='skyblue', edgecolor='navy', linewidth=2,label='July Births')
plt.bar(x + width/2, births_august, width, color='salmon', edgecolor='darkred', linewidth=2,label='August Births')
plt.xlabel("Day of the Week", fontsize=12, fontweight='bold')
plt.ylabel("Number of Births", fontsize=12, fontweight='bold')
plt.title("Weekly Birth Rates Comparison: July vs August", fontsize=14, fontweight='bold')
plt.xticks(x, days, rotation=15)
plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7, color='gray')
plt.legend(title="Month", title_fontsize=12, fontsize=11, loc='upper left', frameon=True, edgecolor='black')
plt.tight_layout()
plt.show()
```
