```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Penguins Data.csv")
df.head(10)
```

```python
df.info()
```

```python
df.columns.tolist()
```

```python
df = df.rename(columns={'Culmen Length (mm)': 'Culmen_Length', 'Culmen Depth (mm)': 'Culmen_Depth','Flipper Length (mm)': 'Flipper_Length','Body Mass (g)': 'Body_Mass'})
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Culmen_Length', y='Body_Mass', hue='Species')
plt.title('Scatter Plot: Culmen Length vs Body Mass')
plt.xlabel('Culmen Length (mm)')
plt.ylabel('Body Mass (g)')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Culmen_Depth', y='Body_Mass', hue='Island')
plt.title('Scatter Plot: Culmen Depth vs Body Mass by Island')
plt.xlabel('Culmen Depth (mm)')
plt.ylabel('Body Mass (g)')
plt.show()
```

```python
sns.pairplot(df, hue='Species', diag_kind='hist', palette='Set2', markers=['o','s','D'])
plt.suptitle('Pair Plot of All Features', y=1.02)
plt.show()
```

```python
area_data = df[['Culmen_Length', 'Body_Mass']]
area_data = area_data.dropna()
area_data = area_data.sort_values(by='Culmen_Length')
area_data.head()
plt.figure(figsize=(8,6))
plt.fill_between(area_data['Culmen_Length'], area_data['Body_Mass'], color='yellow', alpha=0.5)
plt.plot(area_data['Culmen_Length'], area_data['Body_Mass'], color='green')
plt.title('Area Plot: Culmen Length vs Body Mass')
plt.xlabel('Culmen Length (mm)')
plt.ylabel('Body Mass (g)')
plt.show()
```
