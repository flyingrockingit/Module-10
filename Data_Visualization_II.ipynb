```python
import matplotlib.pyplot as plt
months = ["Jan", "Feb", "March", "April", "May", "June"]
sales1 = [100, 125, 150, 200, 145, 170]
sales2 = [90, 130, 140, 180, 160, 190]
plt.figure(figsize=(8, 5))
plt.plot(months, sales1, marker='o', linestyle='-', color='blue', label='Sales 1')
plt.plot(months, sales2, marker='s', linestyle='--', color='red', label='Sales 2')
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales Comparison")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(1, 10, 0.1)
y_linear = (4 * x) + 2
y_quadratic = x**2 - 5*x + 6
plt.plot(x, y_linear, linestyle="--", color='yellow', marker='o', label='y = 4x + 2')
plt.plot(x, y_quadratic, linestyle='-', color='green', label='y = x^2 - 5x + 6')
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Linear and Quadratic Equations")
plt.grid(True)
plt.legend()
plt.show()
```
