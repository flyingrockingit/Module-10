```python
from nltk.chat.util import Chat

reflections = {
    "I am": "You are",
    "I was": "You were",
    "I": "You",
    "I'm": "You're",
    "I'd": "You would",
    "I've": "You have",
    "Me": "You",
    "My": "Your",
    "Mine": "Yours",
    "Myself": "Yourself",
    "You are": "I am",
    "You were": "I was",
    "You're": "I'm",
    "Your": "My"
}

pairs = [
    [
        r"my name is (.*)",
        ["Hello %1, how are you today?"]
    ],
    [
        r"(hi|hello|hey)",
        ["Hello.", "Hey there!"]
    ],
    [
        r"how are you doing\??",
        ["I'm doing good.\nHow about yourself?"]
    ],
    [
        r"what's your name\??",
        ["I am a robot created by AAR 7. You can call me Max."]
    ]
]

def chat():
    print("Hey there peeps.. I am your companion for life created by AAR 7")
    chatbot = Chat(pairs, reflections)
    chatbot.converse()
chat()
```
