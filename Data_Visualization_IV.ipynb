```python
import matplotlib.pyplot as plt

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

bins = [60, 90, 125, 150]

plt.hist(
    [blood_sugar_men, blood_sugar_women],
    bins=bins,
    rwidth=0.95,
    color=['r', 'g'],
    label=['Men', 'Women'],
    edgecolor='black',
    linewidth=1
)

plt.xlabel("Blood Sugar Range")
plt.ylabel("Number of People")
plt.title("Blood Sugar Distribution")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

exam_scores = [45, 67, 89, 55, 76, 90, 34, 56, 78, 88, 92, 61, 73, 84, 95,
               47, 53, 69, 72, 81, 40, 66, 79, 85, 93, 58, 64, 71, 82, 87]

bins = [0, 50, 60, 70, 80, 100]
colors = ['red', 'orange', 'yellow', 'green', 'blue']

counts, bin_edges, patches = plt.hist(
    exam_scores,
    bins=bins,
    rwidth=0.9,
    edgecolor='black',
    label='Students'
)

for patch, color in zip(patches, colors):
    patch.set_facecolor(color)

plt.xlabel("Distribution of the Grading System")
plt.ylabel("Number of Students")
plt.title("Grades Achieved in Trimester 1")
plt.legend()
plt.show()
```
