```python
import matplotlib.pyplot as plt
```

```python
day = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday", "Sunday"]
new_births = [12, 15, 11, 9, 1, 9, 21]
plt.figure(figsize=(8, 5))
plt.plot(day,new_births,marker='D',linestyle='-.',color='purple',linewidth=3,markersize=8,label='New Births')
plt.xlabel("Day of the Week")
plt.ylabel("Number of Births")
plt.title("Birth Rate in July")
plt.grid(True)
plt.legend()
plt.show()
```

```python
day = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday", "Sunday"]
new_births = [17, 5, 2, 11, 1, 8, 29]
plt.figure(figsize=(8, 5))
plt.plot(day,new_births,marker='o',linestyle='--',color='teal',linewidth=2, markersize=10,label='New Births')
plt.xlabel("Day of the Week")
plt.ylabel("Number of Births")
plt.title("Birth Rate in August")
plt.grid(True, linestyle=':', linewidth=0.8)
plt.legend()
plt.show()
```

```python
days = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday", "Sunday"]
births_july = [12, 15, 11, 9, 1, 9, 21]
births_august = [17, 5, 2, 11, 1, 8, 29]
plt.figure(figsize=(10, 6))
plt.plot(days, births_july, marker='o', linestyle='-', color='blue', linewidth=2, markersize=8, label='July')
plt.plot(days, births_august, marker='s', linestyle='--', color='red', linewidth=2, markersize=8, label='August')
plt.xlabel("Day of the Week")
plt.ylabel("Number of Births")
plt.title("Birth Rate Comparison: July vs August")
plt.grid(True)
plt.legend()
plt.show()
```

```python
x_values = list(range(1, 11))
y_values = [6*x**2 + x + 1 for x in x_values]
plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, marker='^', linestyle='-', color='green', linewidth=2, markersize=8,label='y = 6x + x + 1')
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.legend()
plt.title("Quadratic Curve")
plt.grid(True)
plt.show()
```
