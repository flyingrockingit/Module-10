*7.Bivarate Analysis (Relationships)*

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='sex', y='chol', hue='target', palette='Set2')
plt.title("Cholesterol Levels by Sex and Target")
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(df.corr(), annot=True, cmap="Spectral")
plt.title("Correlation Heatmap")
plt.show()
```

```python
numerical_cols = df.select_dtypes(include='number').columns.tolist()
plt.figure(figsize=(20, 15))
for i, col in enumerate(numerical_cols):
    plt.subplot(5, 3, i+1)
    binned_data = pd.cut(df[col], bins=7)
    sns.countplot(x=binned_data, palette="Set3")
    plt.title(f'Count Plot of {col}')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
